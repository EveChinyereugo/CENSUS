## Exploratory Data Analysis

Creating relationship, insight and visualization of our dataset using ---- Univariante ---- Biavariante ---- Multivariant

```python
## Question 1- Which state has the most counties in it?

# Group the data by state and count the number of unique counties in each state

state_counties = Census.groupby('STNAME')['CTYNAME'].nunique()

# Find the state with the maximum number of counties

state_with_most_counties = state_counties.idxmax()

print(state_with_most_counties)
```

```python
# Group the data by state and count the number of unique counties in each state
state_counties = Census.groupby('STNAME')['CTYNAME'].nunique()

# Create a bar chart showing the number of counties in each state
plt.bar(state_counties.index, state_counties.values)


# Add labels and a title to the chart
plt.xlabel('State')
plt.ylabel('Number of Counties')
plt.title('Number of Counties in Each State')
 
# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)

# Display the chart
plt.show()
```

```python
# Group the data by state and count the number of unique counties in each state

state_counties = Census.groupby('STNAME')['CTYNAME'].nunique()

# Sort the states by number of counties and select the top ten

top_states = state_counties.sort_values(ascending=False)[:10]

# Create a bar chart showing the number of counties in each state

plt.bar(top_states.index, top_states.values)

# Add labels and a title to the chart

plt.xlabel('State')

plt.ylabel('Number of Counties')

plt.title('Top Ten States with the Most Counties')

 
# Rotate the x-axis labels for better readability

plt.xticks(rotation=90)

 
# Display the chart

plt.show()
```

```python
# Question 2- The three most populous countries for each state

# Group the data by state and county and sum the population for each county
county_populations = Census.groupby(['STNAME', 'CTYNAME'])['CENSUS2010POP'].sum()

# Group the data by state and sort the counties by population in descending order
state_populations = county_populations.groupby('STNAME').apply(lambda x: x.sort_values(ascending=False)[:3])

# Calculate the total population for each state and sort in descending order
state_total_populations = state_populations.groupby('STNAME').sum().sort_values(ascending=False)

# Get the top three most populous states
top_3_states = state_total_populations.head(3)

# Display the three most populous states
print(top_3_states)
```

```python
# Group the data by state and county and sum the population for each county
county_populations = Census.groupby(['STNAME', 'CTYNAME'])['CENSUS2010POP'].sum()

# Group the data by state and sort the counties by population in descending order
state_populations = county_populations.groupby('STNAME').apply(lambda x: x.sort_values(ascending=False)[:3])

# Sum the populations of the three most populous counties for each state
top_state_populations = state_populations.groupby('STNAME').sum()

# Find the three most populous states
top_three_states = top_state_populations.sort_values(ascending=False)[:3]

# Create a bar chart showing the population of the top three states
plt.bar(top_three_states.index, top_three_states.values)

# Add labels and a title to the chart
plt.xlabel('State')
plt.ylabel('Population')
plt.title('Top Three States by Population')

# Display the chart
plt.show()
```

```python
# Question 3- Which city has the most countries in it

# Group the data by city and county and count the number of counties in each city
city_counties = Census.groupby(['CTYNAME', 'STNAME'])['COUNTY'].count()

# Reset the index of the grouped data
city_counties = city_counties.reset_index()

# Find the city with the maximum number of counties
city_with_most_counties = city_counties.loc[city_counties['COUNTY'].idxmax()]

# Display the city with the most counties
print("City with the most counties:", city_with_most_counties['CTYNAME'])
print("State:", city_with_most_counties['STNAME'])
print("Number of counties:", city_with_most_counties['COUNTY'])
```

```python
# Group the data by city and county and count the number of counties in each city
city_counties = Census.groupby(['CTYNAME', 'STNAME'])['COUNTY'].count()

# Reset the index of the grouped data
city_counties = city_counties.reset_index()

# Sort the data by the count of counties in descending order
sorted_city_counties = city_counties.sort_values('COUNTY', ascending=False)

 # Select the top cities with the most counties
top_cities = sorted_city_counties.head(10)

 # Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(top_cities['CTYNAME'], top_cities['COUNTY'])
plt.xlabel('City')
plt.ylabel('Number of Counties')
plt.title('Cities with the Most Counties')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```

## Observation
District of Columbia has the most countries in it

```python
## Question 4- Which region has the most division in it?

# Group the data by region and division and count the number of divisions in each region
region_divisions = Census.groupby(['REGION', 'DIVISION'])['DIVISION'].count()

# Reset the index of the grouped data and rename the count column
region_divisions = region_divisions.reset_index(name='DIVISION_COUNT')

# Find the region with the most divisions
top_region = region_divisions.sort_values('DIVISION_COUNT', ascending=False).iloc[0]

# Print the result
print(f"The region with the most divisions is Region {top_region['REGION']} (Division {top_region['DIVISION']}) with {top_region['DIVISION_COUNT']} divisions.")
```

```python
# Group the data by region and division and count the number of divisions in each region
region_divisions = Census.groupby(['REGION', 'DIVISION'])['DIVISION'].count()

 
# Reset the index of the grouped data and rename the count column
region_divisions = region_divisions.reset_index(name='DIVISION_COUNT')

 
# Sort the data by division count in descending order
sorted_region_divisions = region_divisions.sort_values('DIVISION_COUNT', ascending=False)

# Plot the data as a bar plot
plt.figure(figsize=(10, 6))
plt.bar(sorted_region_divisions['REGION'].astype(str) + '-' + sorted_region_divisions['DIVISION'].astype(str), sorted_region_divisions['DIVISION_COUNT'])
plt.xlabel('Region-Division')
plt.ylabel('Number of Divisions')
plt.title('Number of Divisions in Each Region-Division')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

```python
# Filter the data for Region 2
region_2_divisions = Census[Census['REGION'] == 2]['DIVISION'].unique()

print("Divisions in Region 2:", region_2_divisions)
```

```python
# Filter the data for Region 2
region_2_divisions = Census[Census['REGION'] == 2]['DIVISION']
 
# Count the occurrences of each division
division_counts = region_2_divisions.value_counts()

# Create a bar plot
plt.figure(figsize=(10, 6))
division_counts.plot(kind='bar')
plt.title('Divisions in Region 2')
plt.xlabel('Division')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
